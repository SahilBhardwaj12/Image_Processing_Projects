```python
"""
Name: Sahil Bhardwaj
Course: Image Processing & Computer Vision
Assignment: Smart Document Scanner
Unit: Image Acquisition & Quality Analysis
Date: ______
"""
print("Smart Document Scanner & Quality Analysis System")
print("This program analyzes sampling and quantization effects on documents")

import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

if not os.path.exists("outputs"):
    os.makedirs("outputs")

img = np.ones((512, 512, 3), dtype=np.uint8) * 200  # Create a gray image
# Add some features to make it interesting
cv2.rectangle(img, (100, 100), (400, 400), (100, 100, 100), -1)
cv2.circle(img, (250, 250), 100, (50, 50, 50), -1)

# Resize
img = cv2.resize(img, (512, 512))

# Convert grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,2,1); plt.title("Original"); plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2); plt.title("Grayscale"); plt.imshow(gray,cmap='gray')
plt.show()

high = gray

med = cv2.resize(gray, (256, 256))

low = cv2.resize(gray, (128, 128))

med_up = cv2.resize(med, (512, 512))
low_up = cv2.resize(low, (512, 512))
plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.title("512x512"); plt.imshow(high,cmap='gray')
plt.subplot(1,3,2); plt.title("256x256"); plt.imshow(med_up,cmap='gray')
plt.subplot(1,3,3); plt.title("128x128"); plt.imshow(low_up,cmap='gray')
plt.show()
def quantize(img, levels):
    factor = 256 // levels
    return (img // factor) * factor

q8  = quantize(gray,256)
q4  = quantize(gray,16)
q2  = quantize(gray,4)
plt.figure(figsize=(10,4))
plt.subplot(1,3,1); plt.title("8-bit"); plt.imshow(q8,cmap='gray')
plt.subplot(1,3,2); plt.title("4-bit"); plt.imshow(q4,cmap='gray')
plt.subplot(1,3,3); plt.title("2-bit"); plt.imshow(q2,cmap='gray')
plt.show()
cv2.imwrite("outputs/high.png",high)
cv2.imwrite("outputs/med.png",med_up)
cv2.imwrite("outputs/low.png",low_up)
cv2.imwrite("outputs/q8.png",q8)
cv2.imwrite("outputs/q4.png",q4)
cv2.imwrite("outputs/q2.png",q2)
```
